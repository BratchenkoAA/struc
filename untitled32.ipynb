```python

def heapify(arr, n, i):
    """Преобразует поддерево с корнем в i в кучу типа Max-Heap."""
    largest = i
    l = 2 * i + 1
    r = 2 * i + 2

    if l < n and arr[i] < arr[l]:
        largest = l

    if r < n and arr[largest] < arr[r]:
        largest = r
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)

def heap_sort(arr):
    """Сортирует массив arr, используя Heapsort."""
    n = len(arr)


    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)


    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        heapify(arr, i, 0)

arr = [25, 10, 30, 5, 15, 20, 35]
heap_sort(arr)
print("Отсортированный массив:", arr)  # Вывод: Отсортированный массив: [5, 10, 15, 20, 25, 30, 35]
```

```python
class Node:
    """
    Представляет узел бинарного дерева.
    """
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


class BinarySearchTree:
    """
    Реализация структуры данных "Бинарное дерево поиска".
    """
    def __init__(self):
        self.root = None

    def insert(self, data):
        """
        Вставляет новый узел с данными ⓃdataⓃ в дерево.
        """
        if self.root is None:
            self.root = Node(data)
        else:
            self._insert_recursive(self.root, data)

    def _insert_recursive(self, node, data):
        """
        Рекурсивная вспомогательная функция для вставки узла.
        """
        if data < node.data:
            if node.left is None:
                node.left = Node(data)
            else:
                self._insert_recursive(node.left, data)
        elif data > node.data:
            if node.right is None:
                node.right = Node(data)
            else:
                self._insert_recursive(node.right, data)
        else:

            return

    def search(self, data):
        """
        Ищет узел с данными ⓃdataⓃ в дереве.
        Возвращает True, если элемент найден, иначе False.
        """
        return self._search_recursive(self.root, data)

    def _search_recursive(self, node, data):
        """
        Рекурсивная вспомогательная функция для поиска узла.
        """
        if node is None:
            return False
        if data == node.data:
            return True
        elif data < node.data:
            return self._search_recursive(node.left, data)
        else:
            return self._search_recursive(node.right, data)

    def delete(self, data):
        """
        Удаляет узел с данными ⓃdataⓃ из дерева.
        """
        self.root = self._delete_recursive(self.root, data)

    def _delete_recursive(self, node, data):
        """
        Рекурсивная вспомогательная функция для удаления узла.
        """
        if node is None:
            return node

        if data < node.data:
            node.left = self._delete_recursive(node.left, data)
        elif data > node.data:
            node.right = self._delete_recursive(node.right, data)
        else:



            if node.left is None:
                return node.right
            elif node.right is None:
                return node.left


            node.data = self._find_min(node.right)


            node.right = self._delete_recursive(node.right, node.data)

        return node

    def _find_min(self, node):
        """
        Находит минимальный элемент в поддереве, начинающемся с ⓃnodeⓃ.
        """
        current = node
        while(current.left is not None):
            current = current.left
        return current.data

    def preorder_traversal(self):
        """
        Выполняет прямой обход дерева (корень, левое поддерево, правое поддерево).
        """
        self._preorder_recursive(self.root)
        print()

    def _preorder_recursive(self, node):
        """
        Рекурсивная вспомогательная функция для прямого обхода.
        """
        if node:
            print(node.data, end=" ")
            self._preorder_recursive(node.left)
            self._preorder_recursive(node.right)

    def inorder_traversal(self):
        """
        Выполняет симметричный обход дерева (левое поддерево, корень, правое поддерево).
        """
        self._inorder_recursive(self.root)
        print()

    def _inorder_recursive(self, node):
        """
        Рекурсивная вспомогательная функция для симметричного
        обхода.
        """
        if node:
            self._inorder_recursive(node.left)
            print(node.data, end=" ")
            self._inorder_recursive(node.right)

    def postorder_traversal(self):
        """
        Выполняет обратный обход дерева (левое поддерево, правое поддерево, корень).
        """
        self._postorder_recursive(self.root)
        print()

    def _postorder_recursive(self, node):
        """
        Рекурсивная вспомогательная функция для обратного обхода.
        """
        if node:
            self._postorder_recursive(node.left)
            self._postorder_recursive(node.right)
            print(node.data, end=" ")


if __name__ == "__main__":
    bst = BinarySearchTree()


    bst.insert(50)
    bst.insert(30)
    bst.insert(20)
    bst.insert(40)
    bst.insert(70)
    bst.insert(60)
    bst.insert(80)

    print("Бинарное дерево:")
    print("Прямой обход:")
    bst.preorder_traversal()  # 50 30 20 40 70 60 80
    print("Симметричный обход:")
    bst.inorder_traversal()   # 20 30 40 50 60 70 80
    print("Обратный обход:")
    bst.postorder_traversal()  # 20 40 30 60 80 70 50

    # Поиск элемента
    print("Поиск элемента 40:", bst.search(40))  # True
    print("Поиск элемента 90:", bst.search(90))  # False

    # Удаление элемента
    print("Удаляем элемент 30")
    bst.delete(30)

    print("Симметричный обход после удаления:")
    bst.inorder_traversal()   # 20 40 50 60 70 80

    print("Удаляем элемент 50 (корень)")
    bst.delete(50)
    print("Симметричный обход после удаления корня:")
    bst.inorder_traversal() # 20 40 60 70 80
```

```python
import random

def heapify(arr, n, i):
    """
    Преобразует поддерево с корнем в i-м индексе в кучу.

    Параметры:
    arr (list): Исходный массив.
    n (int): Размер кучи (часть массива, которую нужно рассматривать как кучу).
    i (int): Индекс корня поддерева.
    """
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2


    if left < n and arr[i] < arr[left]:
        largest = left

    if right < n and arr[largest] < arr[right]:
        largest = right

    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)

def build_heap(arr):
    """
    Строит кучу из массива.

    Параметры:
    arr (list): Исходный массив.
    """
    n = len(arr)
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)

def heap_sort(arr):
    """
    Сортирует массив с использованием сортировки кучей.

    Параметры:
    arr (list): Исходный массив.
    """
    n = len(arr)

    build_heap(arr)
    print("Куча построена:", arr)
    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]


        heapify(arr, i, 0)
        print(f"После извлечения {arr[i]}:", arr)


if __name__ == "__main__":
    arr = [random.randint(1, 100) for _ in range(10)]
    print("Исходный массив:", arr)

    heap_sort(arr)

    print("Отсортированный массив:", arr)
```

```python
def heapify(arr, n, i):
    """Преобразует поддерево с корнем в i в кучу типа Max-Heap."""
    largest = i
    l = 2 * i + 1
    r = 2 * i + 2

    if l < n and arr[i] < arr[l]:
        largest = l

    if r < n and arr[largest] < arr[r]:
        largest = r
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)

def heap_sort(arr):
    """Сортирует массив arr, используя Heapsort."""
    n = len(arr)


    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)


    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        heapify(arr, i, 0)

arr = [25, 10, 30, 5, 15, 20, 35]
heap_sort(arr)
print("Отсортированный массив:", arr)
```

```python
def heapify_min(arr, n, i):
    """Преобразует поддерево с корнем в i в кучу типа Min-Heap."""
    smallest = i
    l = 2 * i + 1
    r = 2 * i + 2

    if l < n and arr[i] > arr[l]:
        smallest = l

    if r < n and arr[smallest] > arr[r]:
        smallest = r

    if smallest != i:
        arr[i], arr[smallest] = arr[smallest], arr[i]
        heapify_min(arr, n, smallest)

def heap_sort_min(arr):
    """Сортирует массив arr, используя Heapsort и Min-кучу (по убыванию)."""
    n = len(arr)


    for i in range(n // 2 - 1, -1, -1):
        heapify_min(arr, n, i)

    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        heapify_min(arr, i, 0)

    arr.reverse()

arr = [25, 10, 30, 5, 15, 20, 35]
heap_sort_min(arr)
print("Отсортированный массив (по убыванию):", arr)
```
