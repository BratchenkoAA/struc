3.2 Алгоритмы поиска и сортировки в одномерных массивах (Python)

1. Реализовать бинарный поиск для отсортированного массива целых чисел в
диапазоне от 1 до 100. Найти число 50.
2. Реализовать сортировку с помощью прямого выбора для массива целых чисел в
диапазоне от 1 до 100.
3. Реализовать сортировку Шелла для массива целых чисел в диапазоне от 1 до 100.

```python
#1
def binary_search(arr, target):
    """
    Функция бинарного поиска элемента в отсортированном массиве.

    Параметры:
        arr (list): Отсортированный массив, в котором выполняется поиск
        target: Искомое значение

    Возвращает:
        int: Индекс найденного элемента или -1, если элемент не найден
    """

    # Инициализация границ поиска
    # left - начало текущего диапазона поиска (индекс первого элемента)
    # right - конец текущего диапазона поиска (индекс последнего элемента)
    left, right = 0, len(arr) - 1

    #  Основной цикл поиска
    # Выполняется пока диапазон поиска не сократится до нуля (left <= right)
    while left <= right:
        #  Вычисление среднего элемента текущего диапазона
        # Используем целочисленное деление для получения индекса середины
        mid = (left + right) // 2

        # 4. Проверка среднего элемента
        if arr[mid] == target:
            #  Элемент найден - возвращаем его индекс
            return mid
        elif arr[mid] < target:
            # Искомый элемент находится в правой половине
            # Сдвигаем левую границу за середину
            left = mid + 1
        else:
            # Искомый элемент находится в левой половине
            # Сдвигаем правую границу перед серединой
            right = mid - 1

    # 5. Элемент не найден
    # Если цикл завершился, а элемент не найден, возвращаем -1
    return -1


# Пример использования функции
if __name__ == "__main__":
    # Создаем отсортированный массив чисел от 1 до 100
    sorted_array = list(range(1, 101))

    # Задаем искомое значение
    target = 50

    # Вызываем функцию бинарного поиска
    result = binary_search(sorted_array, target)

    # Выводим результат
    if result != -1:
        print(f"Число {target} найдено на позиции {result}.")
    else:
        print("Число не найдено.")

# сложность O(log n) т.к. на каждом шаге идет деление на 2


#2
def selection_sort(arr):
    """
    Функция сортировки выбором.
    Сортирует массив путем многократного нахождения минимального элемента
    из неотсортированной части и помещения его в начало.

    Параметры:
        arr (list): Массив для сортировки

    Возвращает:
        list: Отсортированный массив
    """

    #  Внешний цикл по всем элементам массива
    # i - текущая позиция, куда будем ставить минимальный найденный элемент
    for i in range(len(arr)):

        #  Предполагаем, что минимальный элемент - первый в неотсортированной части
        min_idx = i

        #  Внутренний цикл по неотсортированной части массива (справа от i)
        for j in range(i + 1, len(arr)):

            #  Если находим элемент меньше текущего минимума
            if arr[j] < arr[min_idx]:
                # Обновляем индекс минимального элемента
                min_idx = j

        #  После прохода внутреннего цикла меняем местами
        # текущий элемент (i) и найденный минимальный (min_idx)
        arr[i], arr[min_idx] = arr[min_idx], arr[i]

    #  Возвращаем отсортированный массив
    return arr


# Пример использования
if __name__ == "__main__":
    import random

    # Генерируем массив из 10 случайных чисел от 1 до 100
    unsorted_array = random.sample(range(1, 101), 10)
    print("Неотсортированный массив:", unsorted_array)

    # Сортируем копию массива (чтобы сохранить оригинал)
    sorted_array = selection_sort(unsorted_array.copy())
    print("Отсортированный массив:", sorted_array)

# сложность O(n**2) т.к. сложность внутреннего цикла O(n) и внешнего цикла тоже, т.к. на каждои шаге часть неотсортированного массива уменьшается на 1, при переиножении общая сложность O(n**2)


#3
def shell_sort(arr):
    """
    Функция сортировки Шелла.
    Улучшенная версия сортировки вставками, которая сортирует элементы
    с определенным интервалом (gap), постепенно уменьшая его до 1.

    Параметры:
        arr (list): Массив для сортировки

    Возвращает:
        list: Отсортированный массив
    """

    #  Определяем длину массива
    n = len(arr)

    # Начинаем с большого интервала (gap), затем уменьшаем его
    gap = n // 2  # Начальный интервал - половина длины массива

    #  Основной цикл: выполняем, пока интервал больше 0
    while gap > 0:

        #  Сортировка вставками для элементов, отстоящих на gap
        for i in range(gap, n):
            # Сохраняем текущий элемент во временной переменной
            temp = arr[i]

            # Инициализируем индекс для сравнения
            j = i

            #  Сдвигаем элементы, пока не найдем правильное место для temp
            while j >= gap and arr[j - gap] > temp:
                # Сдвигаем элемент на gap позиций вправо
                arr[j] = arr[j - gap]
                # Уменьшаем индекс на gap
                j -= gap

            #  Вставляем сохраненный элемент на правильную позицию
            arr[j] = temp

        #  Уменьшаем интервал вдвое для следующей итерации
        gap = gap // 2

    #  Возвращаем отсортированный массив
    return arr


# Пример использования
if __name__ == "__main__":
    import random

    # Генерируем массив из 10 случайных чисел от 1 до 100
    unsorted_array = random.sample(range(1, 101), 10)
    print("Неотсортированный массив:", unsorted_array)

    # Сортируем копию массива (чтобы сохранить оригинал)
    sorted_array = shell_sort(unsorted_array.copy())
    print("Отсортированный массив:", sorted_array)

#Общая сложность O(n**2)  разделяй и властвуй
```
